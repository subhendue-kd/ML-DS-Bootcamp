```python
##Matplotlib
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

x=np.arange(0,10)
y=np.arange(11,21)
x,y
```

```python
plt.scatter(x,y) 
plt.xlabel('X AXIS')
plt.ylabel('Y AXIS')
plt.title('SCATTER PLOT')
```

```python
y=x**3
y
```

```python
plt.plot(x,y,'y*',linestyle='dashed',linewidth=2,markersize=12)
plt.xlabel('X AXIS')
plt.ylabel('Y AXIS')
plt.title('2D Diagram')
```

```python
plt.subplot(2,2,1)
plt.plot(x,y,'r--')
plt.subplot(2,2,2)
plt.plot(x,y,'g*--')
plt.subplot(2,2,3)
plt.plot(x,y,'bo')
plt.subplot(2,2,4)
plt.plot(x,y,'go')
```

```python
x = np.arange(0, 3*np.pi, 0.1) # agula google korle pawa jai
y = np.sin(x)
plt.plot(x,y)
```

```python
#Bar plot
x1=[2,8,10]
y1=[11,16,9]
x2=[3,9,11]
y2=[6,15,7]
plt.bar(x1,y1)
plt.bar(x2,y2,color='g')
plt.xlabel('X AXIS')
plt.ylabel('Y AXIS')
plt.show()
```

```python
#Histogram
a=np.array([22,87,5,43,56,73,55,54,11,20,51,2,5,79,31,27])
plt.hist(a)
plt.title('Histogram')
plt.show()
```

```python
#Pie chart

labels='Python','C++','Ruby','Java'
sizes=[215,130,245,210]
colors=['gold','yellowgreen','lightcoral','lightblue']
explode=(0.4,0,0.4,0)
plt.pie(sizes,explode=explode,labels=labels,colors=colors,autopct='%1.1f%%')
plt.axis('equal')
plt.show()
```

```python
#Boxplot

data=[np.random.normal(0,std,100) for std in range(1,4)]
plt.boxplot(data,vert=True,patch_artist=False)
plt.show()
```

```python
#Seaborn

import seaborn as sns

df = sns.load_dataset('tips') #  Loading in the data
df.head()

```

```python
df.shape
```

```python
numeric_df = df.select_dtypes(include=['number'])  # Select only numeric columns
correlation_matrix = numeric_df.corr()
correlation_matrix
```

```python
sns.pairplot(df) #pairplot
```

```python
sns.pairplot(df,hue='time') 
```

```python
sns.distplot(df['tip']) #distribution cheak
```

```python
#join plot
sns.jointplot(x='tip',y='total_bill',data=df,kind='hex')
```

```python
sns.countplot(x='sex',data=df)
```

```python
#boxplot and whisker plot
sns.boxplot(x='day', y='total_bill', data=df)
```

```python
sns.boxplot(x='day',y='total_bill',hue='smoker', data=df)
```

```python
sns.violinplot(x='day',y='total_bill', data=df)
```
